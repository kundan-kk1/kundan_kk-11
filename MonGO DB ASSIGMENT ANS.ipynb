```python
Q.5 ans
  
In MongoDB, the methods delete_one(), delete_many(), and drop() are used to remove data or collections from the database. Each of these methods serves a specific purpose:

delete_one():
The delete_one() method is used to delete a single document that matches a specified filter from a collection. If multiple documents match the filter, only the first matching document encountered will be deleted.

Syntax:


db.students.delete_one(filter)
In this example, the delete_one() method will remove the first document from the 'students' collection where the name is "John".

delete_many():
The delete_many() method is used to delete multiple documents that match a specified filter from a collection.

Syntax:

db.collection.delete_many(filter)
filter: Specifies the criteria for selecting the documents to be deleted.


drop():
The drop() method is used to delete an entire collection from the database. Unlike the previous two methods, which remove individual documents, drop() is used to remove the entire collection along with all its documents.

Syntax:

db.collection.drop()
```
