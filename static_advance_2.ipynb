Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with
an example.



ANS   The Probability Mass Function (PMF) and Probability Density Function (PDF) are both mathematical concepts used in probability theory and statistics to describe the probability distribution of a discrete random variable and a continuous random variable, respectively.

Probability Mass Function (PMF):

The PMF is used for discrete random variables, which are variables that can only take on specific, distinct values with certain probabilities.
The PMF gives the probability that a discrete random variable takes on a specific value.
Mathematically, for a discrete random variable X, the PMF is defined as P(X = x), where x represents a specific value that X can take on.
The sum of all probabilities in the PMF equals 1.
Example: Consider rolling a fair six-sided die. The PMF for this scenario would assign a probability of 1/6 to each possible outcome (1, 2, 3, 4, 5, or 6), since each outcome has an equal chance of occurring.
Probability Density Function (PDF):

The PDF is used for continuous random variables, which are variables that can take on any value within a certain range.
The PDF gives the probability density (not the actual probability) of a continuous random variable taking on a specific value within a range.
Unlike the PMF, the PDF does not give the probability of observing a specific value but rather the relative likelihood of observing values within a range.
The area under the PDF curve over a specific interval represents the probability of the random variable falling within that interval.

```python
Q4: Explain the importance of Normal Distribution. Give a few real-life examples of Normal
Distribution.

Ans-:  The normal distribution holds significant importance in various fields due to its widespread applicability and several key properties. Here are some reasons why the normal distribution is important:

Central Limit Theorem (CLT): The normal distribution is a cornerstone of the Central Limit Theorem, which states that the distribution of the sample means of a large number of independent, identically distributed random variables approaches a normal distribution, regardless of the original distribution of the variables. This property makes the normal distribution invaluable in statistical inference and hypothesis testing.

Mathematical Simplicity: The mathematical properties of the normal distribution, such as its symmetric bell-shaped curve, make it analytically tractable. This simplicity facilitates mathematical modeling, computation, and statistical analysis in various fields.

Statistical Inference: Many statistical methods, such as confidence intervals, hypothesis testing, and regression analysis, rely on assumptions of normality. The normal distribution serves as a reference distribution for these methods, allowing researchers to make valid statistical inferences about population parameters based on sample data.

Predictive Modeling: In fields such as finance, economics, and engineering, predictive models often assume that the errors or residuals follow a normal distribution. This assumption allows practitioners to make accurate predictions and assess the uncertainty associated with their models.
```
