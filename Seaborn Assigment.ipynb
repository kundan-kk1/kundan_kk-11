Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
Ans-:    Line Plot: A line plot is used to visualize the change of a variable over time or another continuous variable. It can be used to show trends, patterns, and relationships between variables.
Scatter Plot: A scatter plot is used to visualize the relationship between two variables. It can be used to identify correlations, outliers, and clusters in the data.
Bar Plot: A bar plot is used to compare the values of a categorical variable. It can be used to show the distribution of data across different categories, and to compare the frequency or count of each category.
Box Plot: A box plot is used to visualize the distribution of a variable across different categories. It can be used to identify outliers, skewness, and kurtosis in the data.
Heatmap: A heatmap is used to visualize the correlation or covariance between variables. It can be used to identify patterns, clusters, and relationships between variables.
Here's an example of how to create each of these plots using Seaborn:

import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Line plot
sns.lineplot(x="time", y="total_bill", data=tips)
plt.title("Total Bill Over Time")
plt.xlabel("Time")
plt.ylabel("Total Bill")
plt.show()

# Scatter plot
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Tip vs Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()

# Bar plot
sns.countplot(x="day", data=tips)
plt.title("Number of Tips by Day")
plt.xlabel("Day")
plt.ylabel("Count")
plt.show()

# Box plot
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()

# Heatmap
corr = tips.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
These are just a few examples of the many types of plots that can be created using Seaborn. The library offers a wide range of customization options, allowing users to tailor their plots to their specific needs and preferences.
```

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
Ans-:   

import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Plot a line plot for different events and regions
plt.figure(figsize=(10, 5))

# Plot the signal for the "face" event and "neuroskill" region
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri[fmri["event"] == "face"][fmri["region"] == "neuroskill"])
plt.title("Face Event and Neuroskill Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event")
plt.show()

# Plot the signal for the "house" event and "neuroskill" region
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri[fmri["event"] == "house"][fmri["region"] == "neuroskill"])
plt.title("House Event and Neuroskill Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event")
plt.show()

# Plot the signal for the "face" event and "neuroskill01" region
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri[fmri["event"] == "face"][fmri["region"] == "neuroskill01"])
plt.title("Face Event and Neuroskill01 Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event")
plt.show()

# Plot the signal for the "house" event and "neuroskill01" region
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri[fmri["event"] == "house"][fmri["region"] == "neuroskill01"])
plt.title("House Event and Neuroskill01 Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event")
plt.show()
This code will create two line plots, one for the "face" event and "neuroskill" region, and another for the "house" event and "neuroskill" region. The line plots will show the change of the "signal" variable over time ("timepoint") for each event and region.

Is this conversation helpful so far?


```
